```python
colors=[1,2,4,root.kGreen+2,root.kOrange+4,root.kMagenta,root.kGray,root.kGray,root.kBlack,root.kCyan,root.kOrange,root.kViolet,root.kCyan,root.kGreen,root.kOrange]
file_path="./"
file_names=["2d_mass_pt_03.root","2d_mass_pt_cand.root","2d_mass_pt_05t.root","sim_2d.root"]#,"m_ee_DCA_535.root"],"m_ee_DCA_536.root"]
ee_hist_names=["eemass_default_cent0_sect0","eemass_default_cent1_sect0","eemass_default_cent2_sect0","eemass_default_cent3_sect0","eemass_default_cent4_sect0"]
eeg_hist_names=["eegmass_default_cent0_sect0","eegmass_default_cent1_sect0","eegmass_default_cent2_sect0","eegmass_default_cent3_sect0","eegmass_default_cent4_sect0"]
ee_hist_names_unw=["eemass_unw_default_cent0_sect0","eemass_unw_default_cent1_sect0","eemass_unw_default_cent2_sect0","eemass_unw_default_cent3_sect0","eemass_unw_default_cent4_sect0"]
eeg_hist_names_unw=["eegmass_unw_default_cent0_sect0","eegmass_unw_default_cent1_sect0","eegmass_unw_default_cent2_sect0","eegmass_unw_default_cent3_sect0","eegmass_unw_default_cent4_sect0"]
hist_names = [ee_hist_names,eeg_hist_names,ee_hist_names_unw,eeg_hist_names_unw]
```

```python
hists_read = []

for iFile in range(len(file_names)):
    infile = root.TFile.Open(file_path+file_names[iFile], "read")
    hists_read_file = []
    for itype in range(len(hist_names)):
        hist_read_type = []
        for iname in hist_names[itype]: 
            hist_read_type.append(infile.Get(iname))
            try: hist_read_type[-1].SetDirectory(root.nullptr); hist_read_type[-1].SetName(hist_read_type[-1].GetName()+f"_{iFile}")
            except: 
                print(f"Error setting directory for {hist_names[itype]}{0} in file {file_names[iFile]}")
                break

            hist_read_type[-1].SetName(hist_read_type[-1].GetName()+f"{iFile}")

        hists_read_file.append(hist_read_type)
        
    hists_read.append(hists_read_file)

    infile.Close()
```

```python
itype=1
print(f"Number of hists in type {itype}: {len(hists_read[iFile][itype])}")
c1 = root.TCanvas("c1","c1",1400,800)
c1.Divide(3,2)
mean_pt = [ (pt_edges[i]+pt_edges[i+1])*0.5 for i in range(len(pt_edges)-1) ]
ratios = []; ratios_err = []
for ihist in range(len(hists_read[iFile][itype])):
    c1.cd(ihist+1)
    ratios_centr = []; ratios_centr_err = []
    for ipt in range(1,len(pt_edges)):
        ipt_min = hists_read[0][itype][ihist].GetYaxis().FindBin(pt_edges[ipt-1])
        ipt_max = hists_read[0][itype][ihist].GetYaxis().FindBin(pt_edges[ipt]-0.001)
        pt_slice0 = hists_read[0][itype][ihist].ProjectionX(f"pt_slice0_{ihist}_{ipt}", ipt_min, ipt_max)
        pt_slice1 = hists_read[1][itype][ihist].ProjectionX(f"pt_slice1_{ihist}_{ipt}", ipt_min, ipt_max)
        pt_slice0_unw = hists_read[0][2+itype][ihist].ProjectionX(f"pt_slice0_unw_{ihist}_{ipt}", ipt_min, ipt_max)
        pt_slice1_unw = hists_read[1][2+itype][ihist].ProjectionX(f"pt_slice1_unw_{ihist}_{ipt}", ipt_min, ipt_max)
        ratios_centr.append(pt_slice0.Integral(pt_slice0.FindBin(m_eeg_min), pt_slice0.FindBin(m_eeg_max)) / pt_slice1.Integral(pt_slice1.FindBin(m_eeg_min), pt_slice1.FindBin(m_eeg_max)))
        ratios_centr_err.append(ratios_centr[-1] * ((pt_slice0_unw.Integral(pt_slice0.FindBin(m_eeg_min), pt_slice0.FindBin(m_eeg_max))**-0.5)**2 + (pt_slice1_unw.Integral(pt_slice1.FindBin(m_eeg_min), pt_slice1.FindBin(m_eeg_max))**-0.5)**2)**0.5)

    ratios.append(ratios_centr)
    ratios_err.append(ratios_centr_err)

#c1.Draw()    
```
