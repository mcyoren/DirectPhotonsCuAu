````python
def get_clone(infile, hname, newname=None, verbose=True):
    obj = infile.Get(hname)
    if not obj:
        if verbose:
            print(f"Missing histogram: {hname}")
        return None

    hist = obj.Clone(newname if newname else hname + "_clone")
    hist.SetDirectory(root.nullptr)
    return hist
````

````python
hists = {}

for ifile, fname in enumerate(file_names):
    full_name = file_path + fname
    infile = root.TFile.Open(full_name, "READ")

    if not infile or infile.IsZombie():
        print(f"Cannot open file: {full_name}")
        continue

    print(f"Reading file {ifile}: {full_name}")

    hists[ifile] = {}
    hists[ifile]["filename"] = full_name

    # Main systematic TH2 hists
    for kind in hist_kinds:
        hists[ifile][kind] = {}

        for cfg in configs:
            hists[ifile][kind][cfg] = {}

            for icent in cent_bins:
                hname = hist_names[kind][cfg][icent]
                newname = f"{hname}_file{ifile}"

                hists[ifile][kind][cfg][icent] = get_clone(
                    infile, hname, newname=newname
                )

    # Old-style hists
    hists[ifile]["old"] = {}
    for old_key, cent_dict in old_hist_names.items():
        hists[ifile]["old"][old_key] = {}

        for icent, hname in cent_dict.items():
            newname = f"{hname}_file{ifile}"
            hists[ifile]["old"][old_key][icent] = get_clone(
                infile, hname, newname=newname
            )

    # Cutflows
    hists[ifile]["cutflow"] = {}
    for cfg, hname in cutflow_names.items():
        hists[ifile]["cutflow"][cfg] = get_clone(
            infile, hname, newname=f"{hname}_file{ifile}"
        )

    # Event-level hists
    hists[ifile]["event"] = {}
    for hname in event_hist_names:
        hists[ifile]["event"][hname] = get_clone(
            infile, hname, newname=f"{hname}_file{ifile}", verbose=False
        )

    infile.Close()
````

````python
import math
import ctypes
import os

ROOT_KEEPALIVE = []

def keep(obj):
    if obj:
        ROOT_KEEPALIVE.append(obj)
    return obj

def check_obj(obj, name):
    if obj is None:
        print(f"[BAD] {name}: None")
        return False
    try:
        cname = obj.ClassName()
        oname = obj.GetName()
    except Exception as e:
        print(f"[BAD] {name}: invalid ROOT object: {e}")
        return False

    print(f"[OK] {name}: {cname}  {oname}")
    return True

def check_hist(h, name):
    if not check_obj(h, name):
        return False
    print(f"     entries = {h.GetEntries():.3g}")
    print(f"     integral = {h.Integral():.3g}")
    print(f"     nbinsX = {h.GetNbinsX()}, nbinsY = {h.GetNbinsY() if hasattr(h, 'GetNbinsY') else '-'}")
    return True
````

````python
def clone_hist(h, name):
    if not check_obj(h, f"clone source {name}"):
        return None

    hc = h.Clone(name)
    if not check_obj(hc, f"clone result {name}"):
        return None

    hc.SetDirectory(root.nullptr)
    keep(hc)
    return hc


def project_mass_in_pt(hist2d, pt_min, pt_max, name):
    if not check_obj(hist2d, f"project source {name}"):
        return None

    yaxis = hist2d.GetYaxis()
    y1 = yaxis.FindBin(pt_min + 1e-3)
    y2 = yaxis.FindBin(pt_max - 1e-3)

    print(f"project_mass_in_pt: {name}")
    print(f"  pt range = [{pt_min}, {pt_max})")
    print(f"  y bins   = {y1} to {y2}")

    proj = hist2d.ProjectionX(name, y1, y2)
    if not check_obj(proj, f"projection {name}"):
        return None

    proj.SetDirectory(root.nullptr)
    keep(proj)

    print(f"  proj integral = {proj.Integral():.3g}")
    return proj


def integral_range(hist, xmin, xmax):
    if not check_obj(hist, f"integral source {xmin}_{xmax}"):
        return 0.0, 0.0

    b1 = hist.FindBin(xmin + 1e-3)
    b2 = hist.FindBin(xmax - 1e-3)

    val = 0.0
    err2 = 0.0

    for ib in range(b1, b2 + 1):
        val += hist.GetBinContent(ib)
        err2 += hist.GetBinError(ib)**2

    return float(val), math.sqrt(err2)


def scale_bg_to_sideband(fg, bg, xmin, xmax):
    fg_int, fg_err = integral_range(fg, xmin, xmax)
    bg_int, bg_err = integral_range(bg, xmin, xmax)

    print(f"sideband [{xmin}, {xmax}]: FG={fg_int:.3g}, BG={bg_int:.3g}")

    if bg_int <= 0:
        print("  WARNING: BG sideband <= 0, using scale=1")
        return 1.0

    scale = fg_int / bg_int
    print(f"  scale = {scale:.6g}")
    return scale
````

````python
def sum_cent_mb(hists, ifile, kind, cfg, name):
    """Sum centralities for one file and one config."""
    hsum = None

    for icent in cent_bins:
        h = hists[ifile][kind][cfg][icent]
        if not h:
            continue

        if hsum is None:
            hsum = clone_hist(h, name)
            hsum.Reset("ICESM")

        hsum.Add(h)

    return hsum


def sum_cent_mb_all_files(hists, kind, cfg, name):
    """Sum centralities and files."""
    hsum = None

    for ifile in sorted(hists.keys()):
        if not isinstance(ifile, int):
            continue

        hmb = sum_cent_mb(hists, ifile, kind, cfg, f"{name}_file{ifile}")
        if not hmb:
            continue

        if hsum is None:
            hsum = clone_hist(hmb, name)
            hsum.Reset("ICESM")

        hsum.Add(hmb)

    return hsum

    
````

````python
def get_ee_yield(h_fg_2d, h_bg_2d, pt_min, pt_max, cfg):
    h_fg = project_mass_in_pt(
        h_fg_2d,
        pt_min,
        pt_max,
        f"h_ee_fg_{cfg}_pt{pt_min:g}_{pt_max:g}"
    )

    h_bg = project_mass_in_pt(
        h_bg_2d,
        pt_min,
        pt_max,
        f"h_ee_bg_{cfg}_pt{pt_min:g}_{pt_max:g}"
    )

    if h_fg is None or h_bg is None:
        print("get_ee_yield: missing projection")
        return None

    scale = scale_bg_to_sideband(
        h_fg,
        h_bg,
        windows["ee_norm"][0],
        windows["ee_norm"][1]
    )

    h_bg_scaled = clone_hist(
        h_bg,
        f"h_ee_bg_scaled_{cfg}_pt{pt_min:g}_{pt_max:g}"
    )
    h_bg_scaled.Scale(scale)

    h_sub = clone_hist(
        h_fg,
        f"h_ee_sub_{cfg}_pt{pt_min:g}_{pt_max:g}"
    )
    h_sub.Add(h_bg_scaled, -1.0)

    y_fg, e_fg = integral_range(h_fg, *windows["ee_signal"])
    y_bg, e_bg = integral_range(h_bg_scaled, *windows["ee_signal"])
    y_sub, e_sub = integral_range(h_sub, *windows["ee_signal"])

    print("EE yield:")
    print(f"  FG signal = {y_fg:.3g}")
    print(f"  BG signal = {y_bg:.3g}")
    print(f"  SUB       = {y_sub:.3g} +/- {e_sub:.3g}")

    return {
        "h_fg": h_fg,
        "h_bg": h_bg_scaled,
        "h_sub": h_sub,
        "scale": scale,
        "yield": y_sub,
        "yield_err": e_sub,
        "fg_int": y_fg,
        "bg_int": y_bg,
    }
````

````python
def safe_root_name(s):
    s = str(s)
    for ch in [".", "-", "+", " ", "/", "[", "]", "(", ")", ","]:
        s = s.replace(ch, "_")
    return s


def fit_one_eeg_hist_test(h_sub, name="test_fit"):
    if not check_hist(h_sub, f"{name} input"):
        return None

    if h_sub.Integral() <= 0:
        print(f"[BAD] {name}: histogram integral <= 0, skip fit")
        return None

    tag = safe_root_name(f"{name}_{len(ROOT_KEEPALIVE)}")

    amp0 = max(h_sub.GetMaximum(), 1.0)
    mean0 = 0.135
    sigma0 = 0.018

    f = root.TF1(f"f_{tag}", "gaus(0)+pol2(3)", 0.05, 0.78)

    if not check_obj(f, f"TF1 f_{tag}"):
        return None

    root.SetOwnership(f, False)
    keep(f)

    f.SetParameters(amp0, mean0, sigma0, 0.0, 0.0, 0.0)
    f.SetParLimits(1, 0.10, 0.17)
    f.SetParLimits(2, 0.006, 0.050)

    print("Before fit:")
    print("  amp0 =", amp0)
    print("  mean0 =", mean0)
    print("  sigma0 =", sigma0)

    res = h_sub.Fit(f, "RQS0")

    print("After fit:")

    if not check_obj(f, f"TF1 after fit {tag}"):
        print("[BAD] Fit killed TF1 pointer")
        return None

    status = int(res.Status()) if res else -999

    print("  status =", status)
    print("  mean =", f.GetParameter(1))
    print("  sigma =", f.GetParameter(2))
    print("  chi2/ndf =", f.GetChisquare(), "/", f.GetNDF())

    return f
````

````python
def pol2_subtracted_yield(h, f, xmin, xmax):
    if not check_hist(h, "pol2 yield hist"):
        return 0.0, 0.0, 0.0, 0.0

    if f is None or not check_obj(f, "pol2 yield fit"):
        print("[WARNING] No valid fit. Return raw integral without pol2 subtraction.")
        raw, raw_err = integral_range(h, xmin, xmax)
        return raw, raw_err, raw, 0.0

    pol2 = root.TF1(f"pol2_from_{safe_root_name(f.GetName())}_{len(ROOT_KEEPALIVE)}", "pol2", 0.05, 0.78)
    root.SetOwnership(pol2, False)
    keep(pol2)

    pol2.SetParameters(f.GetParameter(3), f.GetParameter(4), f.GetParameter(5))

    b1 = h.FindBin(xmin + 1e-3)
    b2 = h.FindBin(xmax - 1e-3)

    raw = 0.0
    bg = 0.0
    sig = 0.0
    err2 = 0.0

    for ib in range(b1, b2 + 1):
        x = h.GetBinCenter(ib)
        y = h.GetBinContent(ib)
        b = pol2.Eval(x)

        raw += y
        bg += b
        sig += y - b
        err2 += h.GetBinError(ib)**2

    err = math.sqrt(err2)

    print("pol2-subtracted yield:")
    print(f"  raw = {raw:.3g}")
    print(f"  pol2 = {bg:.3g}")
    print(f"  signal = {sig:.3g} +/- {err:.3g}")

    return sig, err, raw, bg
````

````python
def sum_hists_list(hist_list, name):
    hsum = None

    for h in hist_list:
        if h is None:
            continue

        if hsum is None:
            hsum = clone_hist(h, name)
            hsum.Reset("ICESM")

        hsum.Add(h)

    if hsum is None:
        print(f"[BAD] sum_hists_list: could not build {name}")
        return None

    print(f"[OK] built {name}: integral={hsum.Integral():.6g}")
    return hsum


def build_mb_one_file(hists, ifile, kind, cfg):
    hs = []
    for icent in cent_bins:
        h = hists[ifile][kind][cfg][icent]
        hs.append(h)

    return sum_hists_list(hs, f"mb_file{ifile}_{kind}_{cfg}")


def build_mb_all_files(hists, kind, cfg):
    hs = []

    for ifile in sorted(hists.keys()):
        if not isinstance(ifile, int):
            continue

        h_mb_file = build_mb_one_file(hists, ifile, kind, cfg)
        if h_mb_file is not None:
            hs.append(h_mb_file)

    return sum_hists_list(hs, f"mb_all_{kind}_{cfg}")


def build_mb_set(hists, cfg):
    mb = {}

    mb["fg_ee"] = build_mb_all_files(hists, "fg2d_eemass_reco", cfg)
    mb["bg_ee"] = build_mb_all_files(hists, "bg2d_eemass_reco", cfg)
    mb["fg_eeg"] = build_mb_all_files(hists, "fg2d_eegmass_reco", cfg)
    mb["bg_eeg"] = build_mb_all_files(hists, "bg2d_eegmass_reco", cfg)

    return mb
````

````python
def add_ratio_columns(df):
    df = df.copy()

    # eeg / ee
    df["eeg_over_ee"] = float("nan")
    df["eeg_over_ee_err"] = float("nan")

    # ee / eeg
    df["ee_over_eeg"] = float("nan")
    df["ee_over_eeg_err"] = float("nan")

    for i, row in df.iterrows():
        y_ee = row["y_ee"]
        e_ee = row["e_ee"]
        y_eeg = row["y_eeg"]
        e_eeg = row["e_eeg"]

        if y_ee > 0 and y_eeg > 0:
            r1 = y_eeg / y_ee
            r1_err = r1 * math.sqrt((e_eeg / y_eeg)**2 + (e_ee / y_ee)**2)

            r2 = y_ee / y_eeg
            r2_err = r2 * math.sqrt((e_ee / y_ee)**2 + (e_eeg / y_eeg)**2)

            df.loc[i, "eeg_over_ee"] = r1
            df.loc[i, "eeg_over_ee_err"] = r1_err
            df.loc[i, "ee_over_eeg"] = r2
            df.loc[i, "ee_over_eeg_err"] = r2_err

    return df
````

````python

# -----------------------------------------------------------------------------
# pi0 peak extraction variations for systematic uncertainty
# These variations use the same data histograms as the central conversion cut
# and DO NOT vary the simulation correction.  They only change the pi0/eeg
# extraction procedure in data.
#
# Added configs:
#   pi0_int021  : larger counting window, m_eeγ < 0.21 GeV/c^2
#   pi0_pol1    : Gaussian + pol1 instead of Gaussian + pol2
#   pi0_fit035  : smaller fit window, upper fit bound 0.35 GeV/c^2
# -----------------------------------------------------------------------------

pi0_extraction_variations = [
    {
        "out_cfg": "pi0_int021",
        "base_cfg": "conv_tight",
        "label": "pi0 count window 0.09-0.21",
        "eeg_signal": (0.09, 0.21),
        "fit_xmin": 0.05,
        "fit_xmax": 0.78,
        "pol_order": 2,
    },
    {
        "out_cfg": "pi0_pol1",
        "base_cfg": "conv_tight",
        "label": "pi0 fit Gaussian+pol1",
        "eeg_signal": windows["eeg_signal"],
        "fit_xmin": 0.05,
        "fit_xmax": 0.78,
        "pol_order": 1,
    },
    {
        "out_cfg": "pi0_fit035",
        "base_cfg": "conv_tight",
        "label": "pi0 fit range upper 0.35",
        "eeg_signal": windows["eeg_signal"],
        "fit_xmin": 0.05,
        "fit_xmax": 0.35,
        "pol_order": 2,
    },
]


def fit_one_eeg_hist_test_var(h_sub, name="test_fit", fit_xmin=0.05, fit_xmax=0.78, pol_order=2):
    """Fit pi0 peak with Gaussian + polynomial background.

    The default reproduces fit_one_eeg_hist_test(): gaus + pol2, 0.05-0.78.
    The variation options are used only for the pi0 extraction systematic.
    """
    if not check_hist(h_sub, f"{name} input"):
        return None

    if h_sub.Integral() <= 0:
        print(f"[BAD] {name}: histogram integral <= 0, skip fit")
        return None

    tag = safe_root_name(f"{name}_{len(ROOT_KEEPALIVE)}")

    amp0 = max(h_sub.GetMaximum(), 1.0)
    mean0 = 0.135
    sigma0 = 0.018

    if pol_order == 0:
        formula = "gaus(0)+pol0(3)"
        npar = 4
    elif pol_order == 1:
        formula = "gaus(0)+pol1(3)"
        npar = 5
    elif pol_order == 2:
        formula = "gaus(0)+pol2(3)"
        npar = 6
    else:
        raise ValueError(f"Unsupported pol_order={pol_order}; use 0, 1, or 2")

    f = root.TF1(f"f_{tag}", formula, fit_xmin, fit_xmax)

    if not check_obj(f, f"TF1 f_{tag}"):
        return None

    root.SetOwnership(f, False)
    keep(f)

    pars = [amp0, mean0, sigma0, 0.0, 0.0, 0.0]
    for ip in range(npar):
        f.SetParameter(ip, pars[ip])

    f.SetParLimits(1, 0.10, 0.17)
    f.SetParLimits(2, 0.006, 0.050)

    print("Before fit variation:")
    print("  formula =", formula)
    print("  fit range =", fit_xmin, fit_xmax)
    print("  amp0 =", amp0)
    print("  mean0 =", mean0)
    print("  sigma0 =", sigma0)

    res = h_sub.Fit(f, "RQS0")

    print("After fit variation:")

    if not check_obj(f, f"TF1 after fit {tag}"):
        print("[BAD] Fit killed TF1 pointer")
        return None

    status = int(res.Status()) if res else -999

    print("  status =", status)
    print("  mean =", f.GetParameter(1))
    print("  sigma =", f.GetParameter(2))
    print("  chi2/ndf =", f.GetChisquare(), "/", f.GetNDF())

    return f


def poly_subtracted_yield_var(h, f, xmin, xmax, fit_xmin=0.05, fit_xmax=0.78, pol_order=2):
    """Integrate signal window after subtracting the fitted polynomial background.

    This follows the existing pol2_subtracted_yield convention exactly, but allows
    pol1/pol0 and different fit ranges.
    """
    if not check_hist(h, "poly yield hist"):
        return 0.0, 0.0, 0.0, 0.0

    if f is None or not check_obj(f, "poly yield fit"):
        print("[WARNING] No valid fit. Return raw integral without polynomial subtraction.")
        raw, raw_err = integral_range(h, xmin, xmax)
        return raw, raw_err, raw, 0.0

    if pol_order == 0:
        formula = "pol0"
    elif pol_order == 1:
        formula = "pol1"
    elif pol_order == 2:
        formula = "pol2"
    else:
        raise ValueError(f"Unsupported pol_order={pol_order}; use 0, 1, or 2")

    poly = root.TF1(
        f"poly{pol_order}_from_{safe_root_name(f.GetName())}_{len(ROOT_KEEPALIVE)}",
        formula,
        fit_xmin,
        fit_xmax
    )
    root.SetOwnership(poly, False)
    keep(poly)

    for ip in range(pol_order + 1):
        poly.SetParameter(ip, f.GetParameter(3 + ip))

    b1 = h.FindBin(xmin + 1e-3)
    b2 = h.FindBin(xmax - 1e-3)

    raw = 0.0
    bg = 0.0
    sig = 0.0
    err2 = 0.0

    for ib in range(b1, b2 + 1):
        x = h.GetBinCenter(ib)
        y = h.GetBinContent(ib)
        b = poly.Eval(x)

        raw += y
        bg += b
        sig += y - b
        err2 += h.GetBinError(ib)**2

    err = math.sqrt(err2)

    print(f"pol{pol_order}-subtracted yield:")
    print(f"  signal window = [{xmin}, {xmax}]")
    print(f"  raw = {raw:.3g}")
    print(f"  pol{pol_order} = {bg:.3g}")
    print(f"  signal = {sig:.3g} +/- {err:.3g}")

    return sig, err, raw, bg


def one_bin_result_mb_pi0var(out_cfg, base_cfg, mb, pt_min, pt_max,
                            eeg_signal=None, fit_xmin=0.05, fit_xmax=0.78, pol_order=2):
    print("="*80)
    print(f"one_bin_result_mb_pi0var out_cfg={out_cfg}, base_cfg={base_cfg}, pt=[{pt_min},{pt_max}]")
    print(f"  eeg_signal={eeg_signal}, fit=[{fit_xmin},{fit_xmax}], pol_order={pol_order}")

    if eeg_signal is None:
        eeg_signal = windows["eeg_signal"]

    ee_info = get_ee_yield(
        mb["fg_ee"],
        mb["bg_ee"],
        pt_min,
        pt_max,
        out_cfg
    )

    eeg_pre = get_eeg_subtracted_hist_no_fit(
        mb["fg_eeg"],
        mb["bg_eeg"],
        pt_min,
        pt_max,
        out_cfg
    )

    f = fit_one_eeg_hist_test_var(
        eeg_pre["h_sub"],
        f"eeg_mb_{out_cfg}_pt{pt_min:g}_{pt_max:g}",
        fit_xmin=fit_xmin,
        fit_xmax=fit_xmax,
        pol_order=pol_order,
    )

    y_eeg, e_eeg, raw_eeg, poly_eeg = poly_subtracted_yield_var(
        eeg_pre["h_sub"],
        f,
        eeg_signal[0],
        eeg_signal[1],
        fit_xmin=fit_xmin,
        fit_xmax=fit_xmax,
        pol_order=pol_order,
    )

    y_ee = ee_info["yield"]
    e_ee = ee_info["yield_err"]

    eeg_over_ee = y_eeg / y_ee if y_ee != 0 else float("nan")
    ee_over_eeg = y_ee / y_eeg if y_eeg != 0 else float("nan")

    print("FINAL MB PI0 VAR ONE BIN")
    print(f"  Yee        = {y_ee:.6g} +/- {e_ee:.6g}")
    print(f"  Yeeg       = {y_eeg:.6g} +/- {e_eeg:.6g}")
    print(f"  eeg / ee   = {eeg_over_ee:.6g}")
    print(f"  ee / eeg   = {ee_over_eeg:.6g}")

    return {
        "cfg": out_cfg,
        "base_cfg": base_cfg,
        "pt_min": pt_min,
        "pt_max": pt_max,
        "ee": ee_info,
        "eeg_pre": eeg_pre,
        "fit": f,
        "y_ee": y_ee,
        "e_ee": e_ee,
        "y_eeg": y_eeg,
        "e_eeg": e_eeg,
        "raw_eeg": raw_eeg,
        "pol2_eeg": poly_eeg,   # keep existing column name used downstream
        "eeg_over_ee": eeg_over_ee,
        "ee_over_eeg": ee_over_eeg,
        "eeg_signal_low": eeg_signal[0],
        "eeg_signal_high": eeg_signal[1],
        "fit_xmin": fit_xmin,
        "fit_xmax": fit_xmax,
        "pol_order": pol_order,
    }


def run_pt_bins_for_config_pi0var(out_cfg, base_cfg, mb, save_canvases=False,
                                  eeg_signal=None, fit_xmin=0.05, fit_xmax=0.78, pol_order=2):
    rows = []
    results = {}

    for ipt in range(len(pt_bins) - 1):
        pt_min = pt_bins[ipt]
        pt_max = pt_bins[ipt + 1]

        print("\n")
        print("#"*80)
        print(f"Running pi0 extraction variation {out_cfg}: pT [{pt_min}, {pt_max})")

        res = one_bin_result_mb_pi0var(
            out_cfg,
            base_cfg,
            mb,
            pt_min,
            pt_max,
            eeg_signal=eeg_signal,
            fit_xmin=fit_xmin,
            fit_xmax=fit_xmax,
            pol_order=pol_order,
        )
        results[(pt_min, pt_max)] = res

        if save_canvases:
            tag = f"{out_cfg}_mb_pt{str(pt_min).replace('.','p')}_{str(pt_max).replace('.','p')}"
            save_one_bin_canvas(res, tag)

        row = {
            "config": out_cfg,
            "base_config": base_cfg,
            "pt_min": pt_min,
            "pt_max": pt_max,
            "pt_center": 0.5 * (pt_min + pt_max),

            "y_ee": res["y_ee"],
            "e_ee": res["e_ee"],
            "y_eeg": res["y_eeg"],
            "e_eeg": res["e_eeg"],

            "raw_eeg": res["raw_eeg"],
            "pol2_eeg": res["pol2_eeg"],

            "eeg_over_ee": res["eeg_over_ee"],
            "ee_over_eeg": res["ee_over_eeg"],

            "fit_mean": res["fit"].GetParameter(1) if res["fit"] else float("nan"),
            "fit_sigma": abs(res["fit"].GetParameter(2)) if res["fit"] else float("nan"),
            "fit_chi2": res["fit"].GetChisquare() if res["fit"] else float("nan"),
            "fit_ndf": res["fit"].GetNDF() if res["fit"] else -999,

            "eeg_signal_low": res["eeg_signal_low"],
            "eeg_signal_high": res["eeg_signal_high"],
            "fit_xmin": res["fit_xmin"],
            "fit_xmax": res["fit_xmax"],
            "pol_order": res["pol_order"],
        }

        rows.append(row)

    df = pd.DataFrame(rows)
    return df, results


# Run pi0 extraction variations on the central data cut only.
# The simulation correction is not varied: make_rgamma_graph_for_config_fox()
# will map these configs to the same conv_tight simulation correction.
pi0_var_dfs = []
pi0_var_results = {}

for var in pi0_extraction_variations:
    out_cfg = var["out_cfg"]
    base_cfg = var["base_cfg"]

    print("\n\n")
    print("="*100)
    print(f"BUILDING MB HISTOGRAMS FOR PI0 VARIATION {out_cfg} FROM BASE {base_cfg}")
    print("="*100)

    mb_base = build_mb_set(hists, base_cfg)

    df_var, results_var = run_pt_bins_for_config_pi0var(
        out_cfg,
        base_cfg,
        mb_base,
        save_canvases=False,
        eeg_signal=var["eeg_signal"],
        fit_xmin=var["fit_xmin"],
        fit_xmax=var["fit_xmax"],
        pol_order=var["pol_order"],
    )

    pi0_var_dfs.append(df_var)
    pi0_var_results[out_cfg] = results_var

if len(pi0_var_dfs) > 0:
    df_pi0_variations = pd.concat(pi0_var_dfs, ignore_index=True)
    df_all = pd.concat([df_all, df_pi0_variations], ignore_index=True)
    df_all.to_csv(f"{fit_out_dir}/yields_all_configs_mb_with_pi0_variations.csv", index=False)

df_pi0_variations if len(pi0_var_dfs) > 0 else None
````

````python
import numpy as np
import pandas as pd
import os

def graph_to_arrays(g):
    n = g.GetN()

    x = np.array([float(g.GetX()[i]) for i in range(n)])
    y = np.array([float(g.GetY()[i]) for i in range(n)])

    # TGraphErrors
    ex = np.array([float(g.GetEX()[i]) for i in range(n)]) if hasattr(g, "GetEX") else np.zeros(n)
    ey = np.array([float(g.GetEY()[i]) for i in range(n)]) if hasattr(g, "GetEY") else np.zeros(n)

    order = np.argsort(x)
    return x[order], y[order], ex[order], ey[order]

def build_rgamma_summary(
    graphs,
    nominal_key="nominal",
    syst_keys=None,
    x_match_tol=1e-6,
    syst_method="rms",   # "rms", "std", "envelope", "rms_by_groups", "rms_by_groups_asym"
    additive = 0,
):
    if syst_keys is None:
        syst_keys = [k for k in graphs.keys() if k != nominal_key]

    x_nom, y_nom, ex_nom, ey_nom = graph_to_arrays(graphs[nominal_key])

    def group_name(cfg):
        """
        Map config name to systematic group.
        Adjust this mapping if you rename configs.
        """
        if cfg.startswith("ecore"):
            return "ecore"
        if cfg.startswith("pte"):
            return "pt"
        if cfg.startswith("eid"):
            return "eid"
        if cfg.startswith("chi2"):
            return "chi2"
        if cfg.startswith("escale"):
            return "escale"
        if cfg.startswith("pi0"):
            return "fit"
        if cfg.startswith("conv"):
            return "conv"

        return "other"

    def rms(vals):
        vals = np.asarray(vals, dtype=float)
        vals = vals[np.isfinite(vals)]
        if len(vals) == 0:
            return 0.0
        return float(np.sqrt(np.mean(vals**2)))

    def rms_signed(diffs):
        """
        Return down/up RMS relative to nominal.
        diffs = variation - nominal.
        """
        diffs = np.asarray(diffs, dtype=float)
        diffs = diffs[np.isfinite(diffs)]

        up = diffs[diffs > 0]
        down = -diffs[diffs < 0]

        syst_up = rms(up)
        syst_down = rms(down)

        return syst_down, syst_up

    rows = []

    for i in range(len(x_nom)):
        pt = x_nom[i]
        rg_nom = y_nom[i]
        stat = ey_nom[i]
        ex = ex_nom[i]

        variations = []
        variation_cfgs = []

        for cfg in syst_keys:
            if cfg not in graphs:
                continue

            x_var, y_var, ex_var, ey_var = graph_to_arrays(graphs[cfg])

            idx = np.where(np.abs(x_var - pt) < x_match_tol)[0]

            if len(idx) == 0:
                print(f"Missing matching pT={pt} for cfg={cfg}")
                continue

            y = y_var[idx[0]]

            if not np.isfinite(y):
                print(f"Skipping non-finite variation: cfg={cfg}, pt={pt}, y={y}")
                continue

            variations.append(y)
            variation_cfgs.append(cfg)

        variations = np.array(variations, dtype=float)
        variation_cfgs = np.array(variation_cfgs, dtype=object)

        base_row = {
            "pt": pt,
            "pt_low": pt - ex,
            "pt_high": pt + ex,
            "pt_err": ex,

            "rgamma_nominal": rg_nom,
            "stat_nominal": stat,
        }

        # default group component values
        group_components = {
            "syst_ecore": np.nan,
            "syst_pt": np.nan,
            "syst_eid": np.nan,
            "syst_chi2": np.nan,
            "syst_escale": np.nan,
            "syst_fit": np.nan,
            "syst_conv": np.nan,
            "syst_other": np.nan,
        }

        group_components_asym = {}
        for g in ["ecore", "pt", "eid", "chi2", "escale", "fit", "conv", "other"]:
            group_components_asym[f"syst_{g}_down"] = np.nan
            group_components_asym[f"syst_{g}_up"] = np.nan

        if not np.isfinite(rg_nom) or not np.isfinite(stat):
            row = dict(base_row)
            row.update({
                "syst_down": np.nan,
                "syst_up": np.nan,
                "total_down": np.nan,
                "total_up": np.nan,
                "rgamma_mean_variations": np.nan,
                "rgamma_std_variations": np.nan,
                "n_variations": len(variations),
            })
            row.update(group_components)
            row.update(group_components_asym)
            rows.append(row)
            continue

        if len(variations) > 0:
            diffs = variations - rg_nom

            if syst_method == "envelope":
                syst_up = np.max(np.maximum(diffs, 0.0))
                syst_down = np.max(np.maximum(-diffs, 0.0))

            elif syst_method == "std":
                ddof = 1 if len(variations) > 1 else 0
                syst = np.std(variations, ddof=ddof)
                syst_up = syst
                syst_down = syst

            elif syst_method == "rms":
                syst = rms(diffs)
                syst_up = syst
                syst_down = syst

            elif syst_method == "rms_asym":
                syst_down, syst_up = rms_signed(diffs)

            elif syst_method in ["rms_by_groups", "rms_by_groups_asym"]:
                group_diffs = {
                    "ecore": [],
                    "pt": [],
                    "eid": [],
                    "chi2": [],
                    "escale": [],
                    "fit": [],
                    "conv": [],
                    "other": [],
                }

                for cfg, diff in zip(variation_cfgs, diffs):
                    g = group_name(cfg)
                    group_diffs[g].append(diff)

                if syst_method == "rms_by_groups":
                    # Symmetric uncertainty per group from full spread:
                    # sigma = |max - min| / sqrt(12)
                    # Then add independent groups in quadrature.
                    syst2 = 0.0
                
                    for g, vals in group_diffs.items():
                        vals = np.asarray(vals, dtype=float)
                        vals = vals[np.isfinite(vals)]
                
                        if len(vals) >= 2:
                            comp = abs(np.max(vals) - np.min(vals)) / np.sqrt(12.0)
                        elif len(vals) == 1:
                            # With only one variation there is no max-min spread.
                            # Use absolute shift from nominal as fallback.
                            comp = abs(vals[0])
                        else:
                            comp = 0.0
                
                        group_components[f"syst_{g}"] = comp
                        syst2 += comp**2
                
                    syst = np.sqrt(syst2)
                    syst_up = syst
                    syst_down = syst
                else:
                    # Asymmetric uncertainty per group from observed variation range.
                    #
                    # vals are already deviations from nominal:
                    #     vals = R_var - R_nom
                    #
                    # Therefore:
                    #     upward spread   = max(vals) if max(vals) > 0
                    #     downward spread = -min(vals) if min(vals) < 0
                    #
                    # Convert rectangular spread to 1-sigma equivalent using sqrt(12).
                    # Then add independent groups in quadrature.
                    syst_down2 = 0.0
                    syst_up2 = 0.0
                
                    for g, vals in group_diffs.items():
                        vals = np.asarray(vals, dtype=float)
                        vals = vals[np.isfinite(vals)]
                
                        if len(vals) >= 1:
                            vmax = np.max(vals)
                            vmin = np.min(vals)
                
                            comp_up = max(vmax, 0.0) / np.sqrt(12.0)
                            comp_down = max(-vmin, 0.0) / np.sqrt(12.0)
                        else:
                            comp_up = 0.0
                            comp_down = 0.0
                
                        group_components_asym[f"syst_{g}_down"] = comp_down
                        group_components_asym[f"syst_{g}_up"] = comp_up
                
                        # store symmetric average for convenience
                        group_components[f"syst_{g}"] = 0.5 * (comp_down + comp_up)
                
                        syst_down2 += comp_down**2
                        syst_up2 += comp_up**2
                
                    syst_down = np.sqrt(syst_down2)
                    syst_up = np.sqrt(syst_up2)

            else:
                raise ValueError(f"Unknown syst_method: {syst_method}")

            rg_mean_variations = np.mean(variations)
            rg_std_variations = np.std(variations, ddof=1) if len(variations) > 1 else 0.0

        else:
            syst_up = np.nan
            syst_down = np.nan
            rg_mean_variations = np.nan
            rg_std_variations = np.nan

        syst_down = max(additive, syst_down)
        syst_up = max(additive, syst_up)

        row = dict(base_row)
        row.update({
            "syst_down": syst_down,
            "syst_up": syst_up,

            "total_down": np.sqrt(stat**2 + syst_down**2) if np.isfinite(syst_down) else np.nan,
            "total_up": np.sqrt(stat**2 + syst_up**2) if np.isfinite(syst_up) else np.nan,

            "rgamma_mean_variations": rg_mean_variations,
            "rgamma_std_variations": rg_std_variations,
            "n_variations": len(variations),
        })
        row.update(group_components)
        row.update(group_components_asym)

        rows.append(row)

    return pd.DataFrame(rows)
````

\label{ssec:syst_pi0_extraction}

The tagged photon yield $N_{\rm tag}$ is obtained from the
$M_{e^+e^-\gamma}$ distribution after mixed-event background subtraction.  The
remaining distribution is fitted with a Gaussian signal component and a smooth
polynomial background.  The default extraction uses a Gaussian plus second-order
polynomial fit and integrates the signal in the nominal $\pi^0$ mass window.

To estimate the uncertainty associated with the $\pi^0$ peak extraction, the
analysis is repeated with several alternative but reasonable extraction choices.
Only the data extraction procedure is varied; the simulation correction is kept
fixed to the nominal setup.  The variations include an enlarged integration
window for the $M_{e^+e^-\gamma}$ peak, a Gaussian plus first-order polynomial
background description, and a reduced upper bound of the fit range.  The
resulting values of $R_\gamma$ are shown together with the other cut variations
in Fig.~\ref{fig:syst_rgamma_variations}.

For each $p_T$ bin, the systematic uncertainty is calculated as the RMS of the
absolute deviations of the varied results from the nominal value,
\begin{equation}
  \sigma_{\rm syst}(p_T)
  = \sqrt{\frac{1}{N_{\rm var}}\sum_i
    \left(R_{\gamma,i}(p_T)-R_{\gamma,0}(p_T)\right)^2},
\end{equation}
where $R_{\gamma,0}$ is the nominal result and $R_{\gamma,i}$ are the results
obtained with the alternative analysis choices.  Non-finite points are excluded
from the RMS calculation.  The total uncertainty is then obtained by adding the
statistical and systematic uncertainties in quadrature.
