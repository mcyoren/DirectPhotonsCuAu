```python
def make_standalone_reference_graphs(x, ex, y, ey_stat, ey_syst, name):
    g_stat = root.TGraphErrors(len(x))
    g_syst = root.TGraphErrors(len(x))

    for i in range(len(x)):
        g_stat.SetPoint(i, float(x[i]), float(y[i]))
        g_stat.SetPointError(i, 0.0, float(ey_stat[i]))

        g_syst.SetPoint(i, float(x[i]), float(y[i]))
        g_syst.SetPointError(i, float(ex[i]), float(ey_syst[i]))

    g_stat.SetName(f"g_{name}_stat")
    g_syst.SetName(f"g_{name}_syst")

    return g_stat, g_syst
```

```python
def make_group_graphs(Nch, eNch, Y, eY, sY, group, name):
    idx = group["idx"]
    n = len(idx)

    g_stat = root.TGraphErrors(n)
    g_syst = root.TGraphErrors(n)

    for ip, i in enumerate(idx):
        x = float(Nch[i])
        y = float(Y[i])

        # stat: y stat only
        g_stat.SetPoint(ip, x, y)
        g_stat.SetPointError(ip, 0.0, float(eY[i]))

        # syst: x syst + y syst box
        ex_syst = max(0.05 * x, float(eNch[i]))
        g_syst.SetPoint(ip, x, y)
        g_syst.SetPointError(ip, ex_syst, float(sY[i]))

    g_stat.SetName(f"g_stat_{name}")
    g_syst.SetName(f"g_syst_{name}")

    color = group["color"]
    marker = group["marker"]

    g_stat.SetMarkerStyle(marker)
    g_stat.SetMarkerSize(1.15)
    g_stat.SetMarkerColor(color)
    g_stat.SetLineColor(color)

    # open marker style already controls open/filled for ROOT marker codes
    if not group["filled"]:
        g_stat.SetMarkerColor(color)
        g_stat.SetLineColor(color)

    g_syst.SetFillColorAlpha(color, 0.25)
    g_syst.SetLineColor(color)

    return g_stat, g_syst
```

```python
def make_mb_graphs(dNch, dNch_err, dndy, stat, syst_down, syst_up):
    g_stat = root.TGraphErrors(1)
    g_syst = root.TGraphAsymmErrors(1)

    g_stat.SetName("g_mb_stat")
    g_syst.SetName("g_mb_syst")

    # stat point
    g_stat.SetPoint(0, float(dNch), float(dndy))
    g_stat.SetPointError(0, 0.0, float(stat))

    # syst box: x error is dNch uncertainty, y error is asymmetric syst
    g_syst.SetPoint(0, float(dNch), float(dndy))
    g_syst.SetPointError(
        0,
        float(dNch_err), float(dNch_err),
        float(syst_down), float(syst_up)
    )

    return g_stat, g_syst
```

```python
import numpy as np
import ROOT as root

# ------------------------------------------------------------
# p+p direct photon fit from PRC 109, 044912
#
# d^3N / d^2pT dy = A_pp / (1 + (pT/p0)^2)^n
#
# This is the invariant yield:
# (1 / 2pi pT) d^2N / dpT dy
# ------------------------------------------------------------

A_pp = 1.60e-4   # (GeV/c)^-2
p0_pp = 1.45     # GeV/c
n_pp = 3.3

def pp_invariant_yield(pt, A=A_pp, p0=p0_pp, n=n_pp):
    pt = np.asarray(pt, dtype=float)
    return A / (1.0 + (pt / p0)**2)**n


def integrate_pp_yield(pt_min=1.0, pt_max=5.0, n_steps=20000):
    """
    Returns dN_gamma^dir/dy in p+p for pt_min < pT < pt_max.
    """
    pt = np.linspace(pt_min, pt_max, n_steps)
    inv = pp_invariant_yield(pt)

    # dN/dy = integral 2*pi*pT * invariant_yield dpT
    integrand = 2.0 * np.pi * pt * inv

    return np.trapz(integrand, pt)


Ypp_1_5 = integrate_pp_yield(1.0, 5.0)
print("Integrated p+p direct photon yield, 1<pT<5:", Ypp_1_5)


# ------------------------------------------------------------
# Ncoll vs dNch/deta table from PHENIX direct photon paper
# Use this to build a smooth Ncoll(dNch/deta) interpolation/fit.
# ------------------------------------------------------------

dNch_tab = np.array([
    519.2, 225.4, 85.5, 16.4,
    623.9, 414.2, 274.0, 176.8, 109.4, 61.6, 32.0, 16.0, 7.0
], dtype=float)

dNch_err_tab = np.array([
    26.3, 13.2, 8.0, 2.8,
    32.2, 20.2, 14.8, 11.6, 9.1, 7.1, 5.0, 4.0, 2.0
], dtype=float)

Ncoll_tab = np.array([
    770.6, 282.4, 82.6, 12.1,
    951.0, 590.1, 357.2, 207.5, 111.1, 54.1, 24.0, 10.0, 4.0
], dtype=float)

Ncoll_err_tab = np.array([
    79.9, 28.4, 9.3, 3.1,
    98.5, 61.1, 35.5, 21.2, 10.8, 7.9, 6.0, 3.0, 1.0
], dtype=float)


# ------------------------------------------------------------
# Fit Ncoll(dNch/deta) as power law:
# Ncoll = C * (dNch/deta)^beta
#
# This is only for drawing a smooth band.
# ------------------------------------------------------------

x = dNch_tab
y = Ncoll_tab

mask = np.isfinite(x) & np.isfinite(y) & (x > 0) & (y > 0)

logx = np.log(x[mask])
logy = np.log(y[mask])

# log Ncoll = beta * log(dNch) + logC
coeff, cov = np.polyfit(logx, logy, deg=1, cov=True)

beta_ncoll = coeff[0]
logC_ncoll = coeff[1]
C_ncoll = np.exp(logC_ncoll)

print("Ncoll fit:")
print("  C    =", C_ncoll)
print("  beta =", beta_ncoll)


def Ncoll_from_dNch(dnch):
    dnch = np.asarray(dnch, dtype=float)
    return C_ncoll * dnch**beta_ncoll


def rel_unc_Ncoll_fit(dnch):
    """
    Approximate uncertainty from the covariance of the log-log fit.
    """
    dnch = np.asarray(dnch, dtype=float)
    lx = np.log(dnch)

    # coeff order is [beta, logC]
    # Var(logN) = [logx, 1] cov [logx, 1]^T
    var_logN = (
        cov[0, 0] * lx**2
        + 2.0 * cov[0, 1] * lx
        + cov[1, 1]
    )

    var_logN = np.maximum(var_logN, 0.0)
    return np.sqrt(var_logN)


# ------------------------------------------------------------
# Uncertainty on p+p integrated yield.
#
# The paper band includes uncertainty from the p+p fit and low-pT
# extrapolation. If you later digitize the purple band, replace this
# with the digitized relative uncertainty.
# ------------------------------------------------------------

pp_int_rel_unc = 0.45  # choose 25% as starting value; tune to match paper band

# Optional extra Ncoll normalization uncertainty.
# Set to 0 if you want only fit-band uncertainty.
Ncoll_extra_rel_unc = 0.0


# ------------------------------------------------------------
# Build ROOT graphs: central line + uncertainty band
# ------------------------------------------------------------

xmin = 10.0
xmax = 700.0
npts = 300

xgrid = np.logspace(np.log10(xmin), np.log10(xmax), npts)

Ncoll_grid = Ncoll_from_dNch(xgrid)
ygrid = Ypp_1_5 * Ncoll_grid

rel_ncoll_fit = rel_unc_Ncoll_fit(xgrid)

rel_total = np.sqrt(
    pp_int_rel_unc**2
    + rel_ncoll_fit**2
    + Ncoll_extra_rel_unc**2
)

g_pp_ncoll_scaled = root.TGraph(npts)
g_pp_ncoll_scaled_band = root.TGraphAsymmErrors(npts)

for i in range(npts):
    xx = float(xgrid[i])
    yy = float(ygrid[i])
    ey = float(yy * rel_total[i])

    g_pp_ncoll_scaled.SetPoint(i, xx, yy)

    g_pp_ncoll_scaled_band.SetPoint(i, xx, yy)
    g_pp_ncoll_scaled_band.SetPointError(i, 0.0, 0.0, ey, ey)

g_pp_ncoll_scaled.SetName("g_pp_ncoll_scaled")
g_pp_ncoll_scaled_band.SetName("g_pp_ncoll_scaled_band")

# Style similar to PHENIX purple band
g_pp_ncoll_scaled.SetLineColor(root.kMagenta + 1)
g_pp_ncoll_scaled.SetLineWidth(3)
g_pp_ncoll_scaled.SetFillColorAlpha(root.kMagenta + 1, 0.30)

g_pp_ncoll_scaled_band.SetFillColorAlpha(root.kMagenta + 1, 0.30)
g_pp_ncoll_scaled_band.SetLineColor(root.kMagenta + 1)
g_pp_ncoll_scaled_band.SetLineWidth(1)


# ------------------------------------------------------------
# Draw example:
# Draw this after the frame, before data points.
# ------------------------------------------------------------

# g_pp_ncoll_scaled_band.Draw("3 same")
# g_pp_ncoll_scaled.Draw("L same")
```

```python
def make_total_error(stat, syst_down=None, syst_up=None, syst=None):
    if syst is not None:
        return np.sqrt(stat**2 + syst**2)

    if syst_down is not None and syst_up is not None:
        syst_avg = 0.5 * (abs(syst_down) + abs(syst_up))
        return np.sqrt(stat**2 + syst_avg**2)

    return stat


# ------------------------------------------------------------
# Build one combined graph for fitting:
# archive/reference + extra PPG Au+Au + this Cu+Au MB point
# ------------------------------------------------------------

x_fit = []
y_fit = []
ey_fit = []

# 1) Archive/reference points
for i in range(len(Nch_ref)):
    x_fit.append(float(Nch_ref[i]))
    y_fit.append(float(Yield_ref[i]))

    # use total uncertainty for fit
    ey = make_total_error(
        float(eYield_ref[i]),
        syst=float(sYield_ref[i])
    )
    ey_fit.append(float(ey))


# 2) Extra Au+Au PPG points
for i in range(len(dNch_auau_ppg)):
    x_fit.append(float(dNch_auau_ppg[i]))
    y_fit.append(float(y_auau_ppg[i]))

    ey = make_total_error(
        float(y_auau_ppg_stat[i]),
        syst=float(y_auau_ppg_syst[i])
    )
    ey_fit.append(float(ey))


# 3) This Cu+Au MB point
x_fit.append(float(dNch_mb))
y_fit.append(float(dndy_mb))

ey_mb = make_total_error(
    float(dndy_mb_stat),
    syst_down=float(dndy_mb_syst_down),
    syst_up=float(dndy_mb_syst_up)
)
ey_fit.append(float(ey_mb))


x_fit = np.array(x_fit)
y_fit = np.array(y_fit)
ey_fit = np.array(ey_fit)

# remove bad points
mask_fit = (
    np.isfinite(x_fit) &
    np.isfinite(y_fit) &
    np.isfinite(ey_fit) &
    (x_fit > 0) &
    (y_fit > 0) &
    (ey_fit > 0)
)

x_fit = x_fit[mask_fit]
y_fit = y_fit[mask_fit]
ey_fit = ey_fit[mask_fit]


g_all_fit_with_cuau = root.TGraphErrors(len(x_fit))

for i in range(len(x_fit)):
    g_all_fit_with_cuau.SetPoint(i, float(x_fit[i]), float(y_fit[i]))
    g_all_fit_with_cuau.SetPointError(i, 0.0, float(ey_fit[i]))

g_all_fit_with_cuau.SetName("g_all_fit_with_cuau")
```
