```python
import numpy as np

M_PI0 = 0.1349768  # GeV/c^2

def pi0_inv_spectrum(pt):
    return np.interp(pt, pt_cuau_mb, y_cuau_mb, left=0.0, right=0.0)

def pi0_dndpt_spectrum(pt):
    # input pi0 is invariant yield:
    # inv = 1/(2*pi*pT) dN/dpTdy
    return 2.0 * np.pi * pt * pi0_inv_spectrum(pt)

def boost_x(px, py, pz, e, beta):
    gamma = 1.0 / np.sqrt(1.0 - beta * beta)
    px_new = gamma * (px + beta * e)
    e_new = gamma * (e + beta * px)
    return px_new, py, pz, e_new

def run_pi0_decay_mc(n_mc=2_000_000, pt_parent_min=0.0, pt_parent_max=20.0, seed=12345):
    rng = np.random.default_rng(seed)

    # Generate parent pi0 pT uniformly, weight by physical dN/dpT
    pt_pi0 = rng.uniform(pt_parent_min, pt_parent_max, n_mc)
    w_pi0 = pi0_dndpt_spectrum(pt_pi0)

    # Parent pi0 at y=0 and phi=0; pT distribution only, phi symmetry irrelevant
    e_pi0 = np.sqrt(pt_pi0**2 + M_PI0**2)
    beta = pt_pi0 / e_pi0

    # Isotropic two-body decay in pi0 rest frame
    costh = rng.uniform(-1.0, 1.0, n_mc)
    sinth = np.sqrt(1.0 - costh**2)
    phi = rng.uniform(0.0, 2.0*np.pi, n_mc)

    e_star = M_PI0 / 2.0
    p_star = M_PI0 / 2.0

    px1_star = p_star * sinth * np.cos(phi)
    py1_star = p_star * sinth * np.sin(phi)
    pz1_star = p_star * costh

    px2_star = -px1_star
    py2_star = -py1_star
    pz2_star = -pz1_star

    px1, py1, pz1, e1 = boost_x(px1_star, py1_star, pz1_star, e_star, beta)
    px2, py2, pz2, e2 = boost_x(px2_star, py2_star, pz2_star, e_star, beta)

    pt_g1 = np.sqrt(px1**2 + py1**2)
    pt_g2 = np.sqrt(px2**2 + py2**2)

    return pt_pi0, w_pi0, pt_g1, pt_g2
```

```python
pt_edges = np.array(
    df_rgamma_summary["pt_low"].dropna().tolist()
    + [df_rgamma_summary["pt_high"].dropna().iloc[-1]]
)
pt_edges = np.unique(pt_edges)

pt_centers = 0.5 * (pt_edges[:-1] + pt_edges[1:])
pt_widths = pt_edges[1:] - pt_edges[:-1]

N_PI0_DECAY_MC = 3_000_000
PT_PARENT_MIN = 0.0
PT_PARENT_MAX = 20.0

pt_pi0_mc, w_pi0_mc, pt_g1_mc, pt_g2_mc = run_pi0_decay_mc(
    n_mc=N_PI0_DECAY_MC,
    pt_parent_min=PT_PARENT_MIN,
    pt_parent_max=PT_PARENT_MAX,
    seed=12345
)

# Uniform-MC integration normalization:
# integral f(pT) dpT ≈ (range / N) * sum_i f(pT_i)
mc_norm = (PT_PARENT_MAX - PT_PARENT_MIN) / float(N_PI0_DECAY_MC)

# Parent pi0 dN in pT bins
h_pi0_dN_raw, _ = np.histogram(pt_pi0_mc, bins=pt_edges, weights=w_pi0_mc)
h_pi0_dN = mc_norm * h_pi0_dN_raw

# Decay photons dN in pT bins; each pi0 gives two photons with same parent weight
h_gam_dN_1_raw, _ = np.histogram(pt_g1_mc, bins=pt_edges, weights=w_pi0_mc)
h_gam_dN_2_raw, _ = np.histogram(pt_g2_mc, bins=pt_edges, weights=w_pi0_mc)

h_gam_dN = mc_norm * (h_gam_dN_1_raw + h_gam_dN_2_raw)

# Convert dN/bin to invariant yield
pi0_inv_mc = h_pi0_dN / pt_widths / (2.0 * np.pi * pt_centers)
gam_pi0_decay_inv_mc = h_gam_dN / pt_widths / (2.0 * np.pi * pt_centers)

# Input pi0 invariant yield at same pT
pi0_inv_input = pi0_inv_spectrum(pt_centers)

# Missing factor:
# photons from pi0 decay / parent pi0 invariant yield
pi0_decay_gamma_over_pi0_parent = gam_pi0_decay_inv_mc / pi0_inv_input

print("MC norm =", mc_norm)
print("1 / MC norm =", 1.0 / mc_norm)
```

```python
import numpy as np
import ROOT as root

# ------------------------------------------------------------
# p+p direct photon fit from PRC 109, 044912
# d^3N / d^2pT dy = A_pp / (1 + (pT/p0)^2)^n
# This has the same units as invariant yield:
# (1 / 2pi pT) d^2N / dpT dy
# ------------------------------------------------------------

A_pp = 1.60e-4          # (GeV/c)^-2
p0_pp = 1.45            # GeV/c
n_pp = 3.3

def pp_direct_gamma_fit(pt, A=A_pp, p0=p0_pp, n=n_pp):
    pt = np.asarray(pt, dtype=float)
    return A / (1.0 + (pt / p0)**2)**n


# ------------------------------------------------------------
# Ncoll scaling
# Replace this with the Cu+Au MB <Ncoll> value you want to use.
# ------------------------------------------------------------

Ncoll_cuau_mb = 108.0     # TODO: replace with Cu+Au MB <Ncoll>
Ncoll_rel_unc = .1     # optional, e.g. 0.08 for 8% Ncoll uncertainty


# ------------------------------------------------------------
# Uncertainty band
#
# The PRC paper says the band includes uncertainty from the p+p data
# and the unknown true low-pT functional form.
#
# If you do not have the published numerical band, use a relative
# uncertainty function here. You can tune it.
# ------------------------------------------------------------

def pp_fit_rel_unc(pt):
    pt = np.asarray(pt, dtype=float)

    # Example shape:
    # larger uncertainty below 2 GeV/c,
    # smaller uncertainty at high pT.
    rel = 0.7/pt+0.1

    return rel


# ------------------------------------------------------------
# Build smooth TGraph and TGraphAsymmErrors band
# ------------------------------------------------------------

pt_min = 0.8
pt_max = 7.0
n_points = 300

pt_grid = np.linspace(pt_min, pt_max, n_points)

y_pp = pp_direct_gamma_fit(pt_grid)
y_scaled = Ncoll_cuau_mb * y_pp

rel_pp = pp_fit_rel_unc(pt_grid)

# combine p+p fit uncertainty and Ncoll uncertainty
rel_total = np.sqrt(rel_pp**2 + Ncoll_rel_unc**2)

y_low = y_scaled * (1.0 - rel_total)
y_high = y_scaled * (1.0 + rel_total)

# protect log scale
y_low = np.maximum(y_low, 1e-30)

g_pp_scaled = root.TGraph(n_points)
g_pp_scaled_band = root.TGraphAsymmErrors(n_points)

for i, (pt, y, yl, yh) in enumerate(zip(pt_grid, y_scaled, y_low, y_high)):
    g_pp_scaled.SetPoint(i, float(pt), float(y))

    g_pp_scaled_band.SetPoint(i, float(pt), float(y))
    g_pp_scaled_band.SetPointError(
        i,
        0.0,
        0.0,
        float(y - yl),
        float(yh - y)
    )

g_pp_scaled.SetName("g_pp_scaled_fit")
g_pp_scaled_band.SetName("g_pp_scaled_fit_band")

# Orange style similar to paper
g_pp_scaled.SetLineColor(root.kOrange + 7)
g_pp_scaled.SetLineWidth(3)
g_pp_scaled.SetFillColorAlpha(root.kOrange + 7, 0.25)

g_pp_scaled_band.SetFillColorAlpha(root.kOrange + 7, 0.25)
g_pp_scaled_band.SetLineColor(root.kOrange + 7)
g_pp_scaled_band.SetLineWidth(1)


# ------------------------------------------------------------
# Optional draw example
# ------------------------------------------------------------
# Draw band first, then line, then your data points.
#
# g_pp_scaled_band.Draw("3 same")
# g_pp_scaled.Draw("L same")
```

```python
def integrate_dndy_from_invariant_yield(df, pt_min=1.0, pt_max=5.0):
    dNdy = 0.0
    stat2 = 0.0
    syst_down2 = 0.0
    syst_up2 = 0.0
    rows_used = []

    for row in df.itertuples():
        if not np.isfinite(row.gamma_direct):
            continue
        if not np.isfinite(row.pt_low) or not np.isfinite(row.pt_high):
            continue

        lo = max(row.pt_low, pt_min)
        hi = min(row.pt_high, pt_max)
        if hi <= lo:
            continue

        width = hi - lo
        weight = 2.0 * np.pi * row.pt * width
        val = weight * row.gamma_direct
        stat = weight * row.gamma_direct_stat
        syst_down = weight * row.gamma_direct_syst_down
        syst_up = weight * row.gamma_direct_syst_up
        print(f"pt: {row.pt}, val: {val}, stat: {stat}, syst_down: {syst_down}, syst_up: {syst_up}")

        dNdy += val
        stat2 += stat
        syst_down2 += syst_down
        syst_up2 += syst_up
        #stat2 += stat**2
        #syst_down2 += syst_down**2
        #syst_up2 += syst_up**2

        rows_used.append({
            "pt": row.pt,
            "pt_low_used": lo,
            "pt_high_used": hi,
            "width_used": width,
            "gamma_direct_inv": row.gamma_direct,
            "contribution": val,
            "stat_contribution": stat,
            "syst_down_contribution": syst_down,
            "syst_up_contribution": syst_up,
        })

    return {
        "dNdy": dNdy,
        "stat": (stat2),
        "syst_down": (syst_down2),
        "syst_up": (syst_up2),
        #"stat": np.sqrt(stat2),
        #"syst_down": np.sqrt(syst_down2),
        #"syst_up": np.sqrt(syst_up2),
        "rows_used": pd.DataFrame(rows_used),
    }
```
