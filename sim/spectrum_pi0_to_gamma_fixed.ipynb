```python
import numpy as np

# Cu+Au @ 200 GeV, 0-20%
# ratio = gamma_hadron / gamma_pi0

pt_ratio = np.array([
    0.55, 0.60, 0.70, 0.80, 0.90, 1.00,
    1.25, 1.50, 1.75, 2.00, 2.25, 2.50,
    2.75, 3.00, 3.50, 4.00, 4.50, 5.00,
    5.50, 6.00, 6.50, 7.00, 7.50, 8.00,
    8.50, 9.00, 9.50, 9.90
])

ratio_vals = np.array([
    1.09696, 1.11315, 1.13620, 1.14979, 1.15773, 1.16320,
    1.17504, 1.18419, 1.19127, 1.19712, 1.20184, 1.20584,
    1.20926, 1.21221, 1.21708, 1.22092, 1.22365, 1.22579,
    1.22739, 1.22838, 1.22922, 1.22976, 1.22995, 1.23024,
    1.23024, 1.23028, 1.22996, 1.22980
])

def gamma_hadron_over_gamma_pi0(pt):
    return np.interp(pt, pt_ratio, ratio_vals)
```

```python
import numpy as np

M_PI0 = 0.1349768  # GeV/c^2

def pi0_inv_spectrum(pt):
    return np.interp(pt, pt_cuau_mb, y_cuau_mb, left=0.0, right=0.0)

def pi0_dndpt_spectrum(pt):
    # input pi0 is invariant yield:
    # inv = 1/(2*pi*pT) dN/dpTdy
    return 2.0 * np.pi * pt * pi0_inv_spectrum(pt)

def boost_x(px, py, pz, e, beta):
    gamma = 1.0 / np.sqrt(1.0 - beta * beta)
    px_new = gamma * (px + beta * e)
    e_new = gamma * (e + beta * px)
    return px_new, py, pz, e_new

def run_pi0_decay_mc(n_mc=2_000_000, pt_parent_min=0.0, pt_parent_max=20.0, seed=12345):
    rng = np.random.default_rng(seed)

    # Generate parent pi0 pT uniformly, weight by physical dN/dpT
    pt_pi0 = rng.uniform(pt_parent_min, pt_parent_max, n_mc)
    w_pi0 = pi0_dndpt_spectrum(pt_pi0)

    # Parent pi0 at y=0 and phi=0; pT distribution only, phi symmetry irrelevant
    e_pi0 = np.sqrt(pt_pi0**2 + M_PI0**2)
    beta = pt_pi0 / e_pi0

    # Isotropic two-body decay in pi0 rest frame
    costh = rng.uniform(-1.0, 1.0, n_mc)
    sinth = np.sqrt(1.0 - costh**2)
    phi = rng.uniform(0.0, 2.0*np.pi, n_mc)

    e_star = M_PI0 / 2.0
    p_star = M_PI0 / 2.0

    px1_star = p_star * sinth * np.cos(phi)
    py1_star = p_star * sinth * np.sin(phi)
    pz1_star = p_star * costh

    px2_star = -px1_star
    py2_star = -py1_star
    pz2_star = -pz1_star

    px1, py1, pz1, e1 = boost_x(px1_star, py1_star, pz1_star, e_star, beta)
    px2, py2, pz2, e2 = boost_x(px2_star, py2_star, pz2_star, e_star, beta)

    pt_g1 = np.sqrt(px1**2 + py1**2)
    pt_g2 = np.sqrt(px2**2 + py2**2)

    return pt_pi0, w_pi0, pt_g1, pt_g2
```

```python
pt_edges = np.array(
    df_rgamma_summary["pt_low"].dropna().tolist()
    + [df_rgamma_summary["pt_high"].dropna().iloc[-1]]
)
pt_edges = np.unique(pt_edges)

pt_centers = 0.5 * (pt_edges[:-1] + pt_edges[1:])
pt_widths = pt_edges[1:] - pt_edges[:-1]

N_PI0_DECAY_MC = 3_000_000
PT_PARENT_MIN = 0.0
PT_PARENT_MAX = 20.0

pt_pi0_mc, w_pi0_mc, pt_g1_mc, pt_g2_mc = run_pi0_decay_mc(
    n_mc=N_PI0_DECAY_MC,
    pt_parent_min=PT_PARENT_MIN,
    pt_parent_max=PT_PARENT_MAX,
    seed=12345
)

# Uniform-MC integration normalization:
# integral f(pT) dpT ≈ (range / N) * sum_i f(pT_i)
mc_norm = (PT_PARENT_MAX - PT_PARENT_MIN) / float(N_PI0_DECAY_MC)

# Parent pi0 dN in pT bins
h_pi0_dN_raw, _ = np.histogram(pt_pi0_mc, bins=pt_edges, weights=w_pi0_mc)
h_pi0_dN = mc_norm * h_pi0_dN_raw

# Decay photons dN in pT bins; each pi0 gives two photons with same parent weight
h_gam_dN_1_raw, _ = np.histogram(pt_g1_mc, bins=pt_edges, weights=w_pi0_mc)
h_gam_dN_2_raw, _ = np.histogram(pt_g2_mc, bins=pt_edges, weights=w_pi0_mc)

h_gam_dN = mc_norm * (h_gam_dN_1_raw + h_gam_dN_2_raw)

# Convert dN/bin to invariant yield
pi0_inv_mc = h_pi0_dN / pt_widths / (2.0 * np.pi * pt_centers)
gam_pi0_decay_inv_mc = h_gam_dN / pt_widths / (2.0 * np.pi * pt_centers)

# Input pi0 invariant yield at same pT
pi0_inv_input = pi0_inv_spectrum(pt_centers)

# Missing factor:
# photons from pi0 decay / parent pi0 invariant yield
pi0_decay_gamma_over_pi0_parent = gam_pi0_decay_inv_mc / pi0_inv_input

print("MC norm =", mc_norm)
print("1 / MC norm =", 1.0 / mc_norm)
```

```python
def pi0_decay_gamma_over_pi0_parent_func(pt):
    return np.interp(
        pt,
        df_pi0_decay_factor["pt"].values,
        df_pi0_decay_factor["gamma_pi0_decay_over_pi0_parent"].values,
        left=np.nan,
        right=np.nan
    )

mask_rg = (
    np.isfinite(pt_arr) &
    np.isfinite(rgamma_arr) &
    np.isfinite(stat_arr) &
    np.isfinite(syst_down_arr) &
    np.isfinite(syst_up_arr) &
    (pt_arr >= pt_cuau_mb.min()) &
    (pt_arr <= pt_cuau_mb.max())
)

pt_dir = pt_arr[mask_rg]

Rgamma = rgamma_arr[mask_rg]
Rgamma_stat = stat_arr[mask_rg]
Rgamma_syst_down = syst_down_arr[mask_rg]
Rgamma_syst_up = syst_up_arr[mask_rg]

pi0_parent_inv = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb)
pi0_parent_inv_stat = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_stat)
pi0_parent_inv_syst = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_syst)

pi0_to_gamma = np.array([pi0_decay_gamma_over_pi0_parent_func(pt) for pt in pt_dir])
decay_over_pi0gamma = np.array([gamma_hadron_over_gamma_pi0(pt) for pt in pt_dir])

gamma_pi0_decay_inv = pi0_to_gamma * pi0_parent_inv
gamma_pi0_decay_inv_stat = pi0_to_gamma * pi0_parent_inv_stat
gamma_pi0_decay_inv_syst = pi0_to_gamma * pi0_parent_inv_syst

gamma_decay_inv = decay_over_pi0gamma * gamma_pi0_decay_inv
gamma_decay_inv_stat = decay_over_pi0gamma * gamma_pi0_decay_inv_stat
gamma_decay_inv_syst = decay_over_pi0gamma * gamma_pi0_decay_inv_syst

gamma_direct_inv = (Rgamma - 1.0) * gamma_decay_inv

gamma_direct_inv_stat = np.sqrt(
    (gamma_decay_inv * Rgamma_stat)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_stat)**2
)

gamma_direct_inv_syst_down = np.sqrt(
    (gamma_decay_inv * Rgamma_syst_down)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_syst)**2
)

gamma_direct_inv_syst_up = np.sqrt(
    (gamma_decay_inv * Rgamma_syst_up)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_syst)**2
)
```

```python
import numpy as np
import pandas as pd

mask_rg = (
    np.isfinite(pt_arr) &
    np.isfinite(rgamma_arr) &
    np.isfinite(stat_arr) &
    np.isfinite(syst_down_arr) &
    np.isfinite(syst_up_arr) &
    (pt_arr >= pt_cuau_mb.min()) &
    (pt_arr <= pt_cuau_mb.max())
)

pt_dir = pt_arr[mask_rg]

Rgamma = rgamma_arr[mask_rg]
Rgamma_stat = stat_arr[mask_rg]
Rgamma_syst_down = syst_down_arr[mask_rg]
Rgamma_syst_up = syst_up_arr[mask_rg]

# Same-pT approximation for pi0-decay photon denominator used in the double-ratio method
pi0_gamma_inv = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb)
pi0_gamma_inv_stat = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_stat)
pi0_gamma_inv_syst = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_syst)

# MC cocktail ratio: all decay photons / pi0-decay photons
decay_over_pi0gamma = np.array([gamma_hadron_over_gamma_pi0(pt) for pt in pt_dir])

gamma_decay_inv = decay_over_pi0gamma * pi0_gamma_inv
gamma_decay_inv_stat = decay_over_pi0gamma * pi0_gamma_inv_stat
gamma_decay_inv_syst = decay_over_pi0gamma * pi0_gamma_inv_syst

gamma_direct_inv = (Rgamma - 1.0) * gamma_decay_inv

gamma_direct_inv_stat = np.sqrt(
    (gamma_decay_inv * Rgamma_stat)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_stat)**2
)

gamma_direct_inv_syst_down = np.sqrt(
    (gamma_decay_inv * Rgamma_syst_down)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_syst)**2
)

gamma_direct_inv_syst_up = np.sqrt(
    (gamma_decay_inv * Rgamma_syst_up)**2 +
    ((Rgamma - 1.0) * gamma_decay_inv_syst)**2
)

gamma_direct_dndpt = 2.0 * np.pi * pt_dir * gamma_direct_inv
gamma_direct_dndpt_stat = 2.0 * np.pi * pt_dir * gamma_direct_inv_stat
gamma_direct_dndpt_syst_down = 2.0 * np.pi * pt_dir * gamma_direct_inv_syst_down
gamma_direct_dndpt_syst_up = 2.0 * np.pi * pt_dir * gamma_direct_inv_syst_up

df_direct_gamma_old = pd.DataFrame({
    "pt": pt_dir,

    "Rgamma": Rgamma,
    "Rgamma_stat": Rgamma_stat,
    "Rgamma_syst_down": Rgamma_syst_down,
    "Rgamma_syst_up": Rgamma_syst_up,

    "pi0_gamma_inv": pi0_gamma_inv,
    "pi0_gamma_inv_stat": pi0_gamma_inv_stat,
    "pi0_gamma_inv_syst": pi0_gamma_inv_syst,

    "decay_over_pi0gamma": decay_over_pi0gamma,
    "gamma_decay_inv": gamma_decay_inv,

    "gamma_direct": gamma_direct_inv,
    "gamma_direct_stat": gamma_direct_inv_stat,
    "gamma_direct_syst_down": gamma_direct_inv_syst_down,
    "gamma_direct_syst_up": gamma_direct_inv_syst_up,

    "gamma_direct_dndpt": gamma_direct_dndpt,
    "gamma_direct_dndpt_stat": gamma_direct_dndpt_stat,
    "gamma_direct_dndpt_syst_down": gamma_direct_dndpt_syst_down,
    "gamma_direct_dndpt_syst_up": gamma_direct_dndpt_syst_up,
})

# Make sure all relevant columns are numeric
for col in ["pt", "pt_low", "pt_high", "pt_err"]:
    df_rgamma_summary[col] = pd.to_numeric(df_rgamma_summary[col], errors="coerce")

# Robust merge instead of floating-point matching
df_edges = df_rgamma_summary[["pt", "pt_low", "pt_high", "pt_err"]].copy()
df_edges["pt_key"] = df_edges["pt"].round(6)

df_direct_gamma["pt_key"] = df_direct_gamma["pt"].round(6)

df_direct_gamma = df_direct_gamma.drop(
    columns=[c for c in ["pt_low", "pt_high", "pt_err"] if c in df_direct_gamma.columns]
)

df_direct_gamma = df_direct_gamma.merge(
    df_edges[["pt_key", "pt_low", "pt_high", "pt_err"]],
    on="pt_key",
    how="left"
)

df_direct_gamma = df_direct_gamma.drop(columns=["pt_key"])

df_direct_gamma_old
```

```python
def integrate_dndy_from_invariant_yield(df, pt_min=1.0, pt_max=5.0):
    dNdy = 0.0
    stat2 = 0.0
    syst_down2 = 0.0
    syst_up2 = 0.0
    rows_used = []

    for row in df.itertuples():
        if not np.isfinite(row.gamma_direct):
            continue
        if not np.isfinite(row.pt_low) or not np.isfinite(row.pt_high):
            continue

        lo = max(row.pt_low, pt_min)
        hi = min(row.pt_high, pt_max)
        if hi <= lo:
            continue

        width = hi - lo
        weight = 2.0 * np.pi * row.pt * width

        val = weight * row.gamma_direct
        stat = weight * row.gamma_direct_stat
        syst_down = weight * row.gamma_direct_syst_down
        syst_up = weight * row.gamma_direct_syst_up

        dNdy += val
        stat2 += stat**2
        syst_down2 += syst_down**2
        syst_up2 += syst_up**2

        rows_used.append({
            "pt": row.pt,
            "pt_low_used": lo,
            "pt_high_used": hi,
            "width_used": width,
            "gamma_direct_inv": row.gamma_direct,
            "contribution": val,
            "stat_contribution": stat,
            "syst_down_contribution": syst_down,
            "syst_up_contribution": syst_up,
        })

    return {
        "dNdy": dNdy,
        "stat": np.sqrt(stat2),
        "syst_down": np.sqrt(syst_down2),
        "syst_up": np.sqrt(syst_up2),
        "rows_used": pd.DataFrame(rows_used),
    }
```

```python
res_dndy_1_5 = integrate_dndy_from_invariant_yield(
    df_direct_gamma,
    pt_min=1.0,
    pt_max=5.0
)

print("dN/dy direct gamma, 1<pT<5:")
print("value     =", res_dndy_1_5["dNdy"])
print("stat      =", res_dndy_1_5["stat"])
print("syst_down =", res_dndy_1_5["syst_down"])
print("syst_up   =", res_dndy_1_5["syst_up"])

res_dndy_1_5["rows_used"]
```
