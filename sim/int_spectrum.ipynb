```python
def make_standalone_reference_graphs(x, ex, y, ey_stat, ey_syst, name):
    g_stat = root.TGraphErrors(len(x))
    g_syst = root.TGraphErrors(len(x))

    for i in range(len(x)):
        g_stat.SetPoint(i, float(x[i]), float(y[i]))
        g_stat.SetPointError(i, 0.0, float(ey_stat[i]))

        g_syst.SetPoint(i, float(x[i]), float(y[i]))
        g_syst.SetPointError(i, float(ex[i]), float(ey_syst[i]))

    g_stat.SetName(f"g_{name}_stat")
    g_syst.SetName(f"g_{name}_syst")

    return g_stat, g_syst
```

```python
def make_group_graphs(Nch, eNch, Y, eY, sY, group, name):
    idx = group["idx"]
    n = len(idx)

    g_stat = root.TGraphErrors(n)
    g_syst = root.TGraphErrors(n)

    for ip, i in enumerate(idx):
        x = float(Nch[i])
        y = float(Y[i])

        # stat: y stat only
        g_stat.SetPoint(ip, x, y)
        g_stat.SetPointError(ip, 0.0, float(eY[i]))

        # syst: x syst + y syst box
        ex_syst = max(0.05 * x, float(eNch[i]))
        g_syst.SetPoint(ip, x, y)
        g_syst.SetPointError(ip, ex_syst, float(sY[i]))

    g_stat.SetName(f"g_stat_{name}")
    g_syst.SetName(f"g_syst_{name}")

    color = group["color"]
    marker = group["marker"]

    g_stat.SetMarkerStyle(marker)
    g_stat.SetMarkerSize(1.15)
    g_stat.SetMarkerColor(color)
    g_stat.SetLineColor(color)

    # open marker style already controls open/filled for ROOT marker codes
    if not group["filled"]:
        g_stat.SetMarkerColor(color)
        g_stat.SetLineColor(color)

    g_syst.SetFillColorAlpha(color, 0.25)
    g_syst.SetLineColor(color)

    return g_stat, g_syst
```

```python
def make_mb_graphs(dNch, dNch_err, dndy, stat, syst_down, syst_up):
    g_stat = root.TGraphErrors(1)
    g_syst = root.TGraphAsymmErrors(1)

    g_stat.SetName("g_mb_stat")
    g_syst.SetName("g_mb_syst")

    # stat point
    g_stat.SetPoint(0, float(dNch), float(dndy))
    g_stat.SetPointError(0, 0.0, float(stat))

    # syst box: x error is dNch uncertainty, y error is asymmetric syst
    g_syst.SetPoint(0, float(dNch), float(dndy))
    g_syst.SetPointError(
        0,
        float(dNch_err), float(dNch_err),
        float(syst_down), float(syst_up)
    )

    return g_stat, g_syst
```

```python
def make_total_error(stat, syst_down=None, syst_up=None, syst=None):
    if syst is not None:
        return np.sqrt(stat**2 + syst**2)

    if syst_down is not None and syst_up is not None:
        syst_avg = 0.5 * (abs(syst_down) + abs(syst_up))
        return np.sqrt(stat**2 + syst_avg**2)

    return stat


# ------------------------------------------------------------
# Build one combined graph for fitting:
# archive/reference + extra PPG Au+Au + this Cu+Au MB point
# ------------------------------------------------------------

x_fit = []
y_fit = []
ey_fit = []

# 1) Archive/reference points
for i in range(len(Nch_ref)):
    x_fit.append(float(Nch_ref[i]))
    y_fit.append(float(Yield_ref[i]))

    # use total uncertainty for fit
    ey = make_total_error(
        float(eYield_ref[i]),
        syst=float(sYield_ref[i])
    )
    ey_fit.append(float(ey))


# 2) Extra Au+Au PPG points
for i in range(len(dNch_auau_ppg)):
    x_fit.append(float(dNch_auau_ppg[i]))
    y_fit.append(float(y_auau_ppg[i]))

    ey = make_total_error(
        float(y_auau_ppg_stat[i]),
        syst=float(y_auau_ppg_syst[i])
    )
    ey_fit.append(float(ey))


# 3) This Cu+Au MB point
x_fit.append(float(dNch_mb))
y_fit.append(float(dndy_mb))

ey_mb = make_total_error(
    float(dndy_mb_stat),
    syst_down=float(dndy_mb_syst_down),
    syst_up=float(dndy_mb_syst_up)
)
ey_fit.append(float(ey_mb))


x_fit = np.array(x_fit)
y_fit = np.array(y_fit)
ey_fit = np.array(ey_fit)

# remove bad points
mask_fit = (
    np.isfinite(x_fit) &
    np.isfinite(y_fit) &
    np.isfinite(ey_fit) &
    (x_fit > 0) &
    (y_fit > 0) &
    (ey_fit > 0)
)

x_fit = x_fit[mask_fit]
y_fit = y_fit[mask_fit]
ey_fit = ey_fit[mask_fit]


g_all_fit_with_cuau = root.TGraphErrors(len(x_fit))

for i in range(len(x_fit)):
    g_all_fit_with_cuau.SetPoint(i, float(x_fit[i]), float(y_fit[i]))
    g_all_fit_with_cuau.SetPointError(i, 0.0, float(ey_fit[i]))

g_all_fit_with_cuau.SetName("g_all_fit_with_cuau")
```

```python
pt_label = f"{pt_min:g}<p_{{T}}<{pt_max:g} GeV/c"

c_dndy_vs_dNch = root.TCanvas("c_dndy_vs_dNch", "c_dndy_vs_dNch", 850, 700)
c_dndy_vs_dNch.SetLogy()
c_dndy_vs_dNch.SetLogx()

frame = root.TH2F(
    "frame_dndy_vs_dNch",
    f";dN_{{ch}}/d#eta;dN_{{#gamma}}^{{dir}}/dy, {pt_label}",
    100, 1.0, 2e3,
    100, 1e-4, 20.0
)
frame.SetStats(0)
frame.Draw()

# Draw systematic boxes first
for item in ref_graphs:
    item["syst"].Draw("2 same")

# Draw stat markers second
for item in ref_graphs:
    item["stat"].Draw("P same")

#new auau 
# Extra PPG Au+Au 200 systematic boxes
g_auau_ppg_syst.Draw("2 same")
g_auau_ppg_stat.Draw("P same")

# Your MB syst
g_mb_syst.SetFillColorAlpha(root.kMagenta, 0.50)
g_mb_syst.SetLineColor(root.kMagenta)
g_mb_syst.Draw("2 same")

# Your MB stat
g_mb_stat.SetMarkerStyle(33)
g_mb_stat.SetMarkerSize(3.0)
g_mb_stat.SetMarkerColor(root.kMagenta)
g_mb_stat.SetLineColor(root.kMagenta)
g_mb_stat.Draw("P same")

# Legend similar to published plot
leg = root.TLegend(0.15, 0.60, 0.58, 0.88)
leg.SetBorderSize(0)
leg.SetFillStyle(0)
leg.SetTextSize(0.030)

# Avoid duplicate legend entries for Au+Au 200 if filled/open are both present
added = set()
leg.AddEntry(g_auau_ppg_stat, "PHENIX Au+Au 200 GeV, PPG", "p")

for item in ref_graphs:
    name = item["group"]["name"]
    if name in added:
        continue
    leg.AddEntry(item["stat"], name, "p")
    added.add(name)

leg.AddEntry(g_mb_stat, "Cu+Au MB, this analysis", "p")
leg.AddEntry(g_mb_syst, "Cu+Au MB syst.", "f")

leg.Draw()

# Optional label like the plot
lat = root.TLatex()
lat.SetNDC(True)
lat.SetTextSize(0.035)
lat.DrawLatex(0.17, 0.88, f"Direct #gamma ({pt_label})")

x_fit = []
y_fit = []
ey_fit = []

# Archive/reference points
for i in range(len(Nch_ref)):
    x_fit.append(float(Nch_ref[i]))
    y_fit.append(float(Yield_ref[i]))
    ey_fit.append(float(np.sqrt(eYield_ref[i]**2 + 0*sYield_ref[i]**2)))

# Extra Au+Au PPG points
for i in range(len(dNch_auau_ppg)):
    x_fit.append(float(dNch_auau_ppg[i]))
    y_fit.append(float(y_auau_ppg[i]))
    ey_fit.append(float(np.sqrt(y_auau_ppg_stat[i]**2 + 0*y_auau_ppg_syst[i]**2)))

# Your Cu+Au MB point
x_fit.append(float(dNch_mb))
y_fit.append(float(dndy_mb))
ey_fit.append(float(np.sqrt(
    dndy_mb_stat**2 +
    (0.5 * (dndy_mb_syst_down + 0*dndy_mb_syst_up))**2
)))

g_all_fit = root.TGraphErrors(len(x_fit))

for i in range(len(x_fit)):
    g_all_fit.SetPoint(i, x_fit[i], y_fit[i])
    g_all_fit.SetPointError(i, 0.0, ey_fit[i])

f_scale = root.TF1("f_scale", "[0]*pow(x,[1])", 5.0, 1500.0)
f_scale.SetParameters(1e-4, 1.25)
f_scale.SetLineColor(root.kBlack)
f_scale.SetLineStyle(2)
f_scale.SetLineWidth(2)

g_all_fit.Fit(f_scale, "RQ0")
f_scale.Draw("same")

lat.DrawLatex(0.63, 0.24, f"#alpha = {f_scale.GetParameter(1):.2f} #pm {f_scale.GetParError(1):.2f}")

c_dndy_vs_dNch.Draw()
```
