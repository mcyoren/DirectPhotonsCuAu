```python

%run ../real/input/Format.ipynb
import ROOT as root
from array import array
root.gErrorIgnoreLevel = root.kFatal
%jsroot on
root.gStyle.SetOptStat(0)
import ctypes
```

```python
colors=[1,2,4,root.kGreen+2,root.kOrange+4,root.kMagenta,root.kGray,root.kGray,root.kBlack,root.kCyan,root.kOrange,root.kViolet,root.kCyan,root.kGreen,root.kOrange]
file_path="input/"
file_name="narrow.root"#,"m_ee_DCA_535.root"],"m_ee_DCA_536.root"]
```

```python
hist_names = ["h2d_EMCdead_sect0", "h2d_EMCdead_sect1", "h2d_EMCdead_sect2", "h2d_EMCdead_sect3", "h2d_EMCdead_sect4", "h2d_EMCdead_sect5"]
```

```python
input_file = root.TFile(file_path+file_name)
hists = []
for hist_name in hist_names:
    hists.append(input_file.Get(hist_name))
    hists[-1].SetDirectory(0)
input_file.Close()
```

```python
c1 = root.TCanvas("c1", "c1", 1400, 900)
c1.Divide(3,2)
for i in range(6):
    c1.cd(i+1)
    hists[i].SetTitle("EMC sector "+str(i))
    hists[i].GetXaxis().SetRangeUser(0, 36)
    hists[i].GetYaxis().SetRangeUser(0, 72)
    hists[i].GetXaxis().SetTitle("IY")
    hists[i].GetYaxis().SetTitle("IZ")
    hists[i].Draw("colz")
c1.Draw()
c1.SaveAs("output/EMCdead_sect.png")
```
