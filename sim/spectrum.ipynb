```python
import numpy as np

# Cu+Au @ 200 GeV, 0-20%
# ratio = gamma_hadron / gamma_pi0

pt_ratio = np.array([
    0.55, 0.60, 0.70, 0.80, 0.90, 1.00,
    1.25, 1.50, 1.75, 2.00, 2.25, 2.50,
    2.75, 3.00, 3.50, 4.00, 4.50, 5.00,
    5.50, 6.00, 6.50, 7.00, 7.50, 8.00,
    8.50, 9.00, 9.50, 9.90
])

ratio_vals = np.array([
    1.09696, 1.11315, 1.13620, 1.14979, 1.15773, 1.16320,
    1.17504, 1.18419, 1.19127, 1.19712, 1.20184, 1.20584,
    1.20926, 1.21221, 1.21708, 1.22092, 1.22365, 1.22579,
    1.22739, 1.22838, 1.22922, 1.22976, 1.22995, 1.23024,
    1.23024, 1.23028, 1.22996, 1.22980
])

def gamma_hadron_over_gamma_pi0(pt):
    return np.interp(pt, pt_ratio, ratio_vals)
```

```python
import numpy as np
import pandas as pd

# Remove bad rgamma points and require overlap with pi0 spectrum
mask_rg = (
    np.isfinite(pt_arr) &
    np.isfinite(rgamma_arr) &
    np.isfinite(stat_arr) &
    np.isfinite(syst_down_arr) &
    np.isfinite(syst_up_arr) &
    (pt_arr >= pt_cuau_mb.min()) &
    (pt_arr <= pt_cuau_mb.max())
)

pt_dir = pt_arr[mask_rg]
Rgamma = rgamma_arr[mask_rg]
Rgamma_stat = stat_arr[mask_rg]
Rgamma_syst_down = syst_down_arr[mask_rg]
Rgamma_syst_up = syst_up_arr[mask_rg]

# Interpolate pi0 spectrum to Rgamma pT points
pi0_y = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb)
pi0_stat = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_stat)
pi0_syst = np.interp(pt_dir, pt_cuau_mb, y_cuau_mb_syst)

# gamma_decay / pi0 from your function
decay_over_pi0 = np.array([gamma_hadron_over_gamma_pi0(pt) for pt in pt_dir])

# Decay photon spectrum
gamma_decay = decay_over_pi0 * pi0_y
gamma_decay_stat = decay_over_pi0 * pi0_stat
gamma_decay_syst = decay_over_pi0 * pi0_syst

# Direct photon spectrum:
# gamma_dir = gamma_inc - gamma_decay = (Rgamma - 1) * gamma_decay
gamma_direct = (Rgamma - 1.0) * gamma_decay

# Uncertainty propagation
# d gamma_dir / d Rgamma = gamma_decay
# d gamma_dir / d gamma_decay = Rgamma - 1
gamma_direct_stat = np.sqrt(
    (gamma_decay * Rgamma_stat)**2 +
    ((Rgamma - 1.0) * gamma_decay_stat)**2
)

gamma_direct_syst_down = np.sqrt(
    (gamma_decay * Rgamma_syst_down)**2 +
    ((Rgamma - 1.0) * gamma_decay_syst)**2
)

gamma_direct_syst_up = np.sqrt(
    (gamma_decay * Rgamma_syst_up)**2 +
    ((Rgamma - 1.0) * gamma_decay_syst)**2
)

df_direct_gamma = pd.DataFrame({
    "pt": pt_dir,
    "Rgamma": Rgamma,
    "Rgamma_stat": Rgamma_stat,
    "Rgamma_syst_down": Rgamma_syst_down,
    "Rgamma_syst_up": Rgamma_syst_up,

    "pi0_y": pi0_y,
    "pi0_stat": pi0_stat,
    "pi0_syst": pi0_syst,

    "decay_over_pi0": decay_over_pi0,
    "gamma_decay": gamma_decay,

    "gamma_direct": gamma_direct,
    "gamma_direct_stat": gamma_direct_stat,
    "gamma_direct_syst_down": gamma_direct_syst_down,
    "gamma_direct_syst_up": gamma_direct_syst_up,
})

df_direct_gamma
```

```python
def integrate_dndy_from_invariant_yield(df, pt_min=1.0, pt_max=5.0):
    dNdy = 0.0
    stat2 = 0.0
    syst_down2 = 0.0
    syst_up2 = 0.0

    rows_used = []

    for row in df.itertuples():
        if not np.isfinite(row.gamma_direct):
            continue
        if not np.isfinite(row.pt_low) or not np.isfinite(row.pt_high):
            continue

        # overlap of this pT bin with requested range
        lo = max(row.pt_low, pt_min)
        hi = min(row.pt_high, pt_max)

        if hi <= lo:
            continue

        width = hi - lo
        pt = row.pt

        weight = 2.0 * np.pi * pt * width

        val = weight * row.gamma_direct
        stat = weight * row.gamma_direct_stat
        syst_down = weight * row.gamma_direct_syst_down
        syst_up = weight * row.gamma_direct_syst_up

        dNdy += val
        stat2 += stat**2
        syst_down2 += syst_down**2
        syst_up2 += syst_up**2

        rows_used.append({
            "pt": pt,
            "pt_low_used": lo,
            "pt_high_used": hi,
            "width_used": width,
            "gamma_direct": row.gamma_direct,
            "contribution": val,
            "stat_contribution": stat,
            "syst_down_contribution": syst_down,
            "syst_up_contribution": syst_up,
        })

    return {
        "dNdy": dNdy,
        "stat": np.sqrt(stat2),
        "syst_down": np.sqrt(syst_down2),
        "syst_up": np.sqrt(syst_up2),
        "rows_used": pd.DataFrame(rows_used),
    }
```
